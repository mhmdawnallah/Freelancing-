```python
import requests as req
from bs4 import BeautifulSoup

r = req.get("https://clutch.co/seo-firms")
soup = BeautifulSoup(r.content,'html.parser')
page_nr = soup.find_all('a',{'class':'page-link'})[-1]['data-page']
all = soup.findAll('li', attrs={'class':'provider provider-row sponsor'})
```

### Scarpping SEO Agency Name

```python
ls=[]
base_url="https://clutch.co/seo-firms?page="
for page in range(0,int(2),1):
    r = req.get(base_url+str(page)+".html")
    soup = BeautifulSoup(r.content,'html.parser')
    for div in all:
        d={}
        d['Title'] = div.find_all('a',attrs={'data-link_text':'Profile Title'})[0].text.strip()
        d['Tagline'] = div.find_all('p',attrs={'class':'company_info__wrap tagline'})[0].text.strip()
        d['Rating'] = div.find_all('span',attrs={'class':'company_info__wrap tagline'})[0]
        try:
            d['Reviews'] = div.find_all('a',attrs={'data_link_text':'Reviews Count'})[0].text
        except:
            pass
        d['Quote'] = div.find_all('blackquote',attrs={'class':'blockquote-in-module'})[0][0].text
        
        ls.append(d)
```

```python
import pandas as pd
df = pd.DataFrame(ls)
df
```
